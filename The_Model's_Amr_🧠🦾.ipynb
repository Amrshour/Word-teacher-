```python
from textblob import Word
```

```python
print("ðŸ‘‹ Hello! I'm the Word Teacher.")
print("Type One word, and I will analyze it with NIP!")

while True:
      user_word=input("\n Enter a word  (or 'quit' to stop)").lower()
      if user_word=='quit':
          print("Goodbye")
          break
      w=Word(user_word)
      corrected=w.spellcheck()[0][0]
      lema=Word(corrected).lemmatize('v')
      print(f"\nðŸ”Ž NLP Analysis for '{user_word}':")
      print(f"âœ… Corrected spelling: {corrected}")
      print(f"ðŸŒ± Lemma (base form): {lema}")

```
